# Dengue outbreaks in Kolkata

Dengue outbreaks in Kolkata analyzed based on temperature and humidity during different months of the year. 

### Data preparation (synthetic)

```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Generate synthetic data
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
temperature = np.random.uniform(15, 35, 12)  # Random temperatures between 15°C and 35°C
humidity = np.random.uniform(40, 90, 12)  # Random humidity levels between 40% and 90%
dengue_cases = np.random.randint(50, 500, 12)  # Random number of dengue cases

# Create DataFrame
data = {
    'Month': months,
    'Temperature (°C)': temperature,
    'Humidity (%)': humidity,
    'Dengue Cases': dengue_cases
}
df = pd.DataFrame(data)

# Save the dataset to a CSV file
df.to_csv('dengue_outbreaks_kolkata.csv', index=False)
print("Dataset generated and saved as 'dengue_outbreaks_kolkata.csv'")
```

# Plotting of the chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('dengue_outbreaks_kolkata.csv')

# Create figure and axis objects
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot temperature on primary y-axis
ax1.plot(df['Month'], df['Temperature (°C)'], marker='o', label='Temperature (°C)', color='tab:orange')
ax1.set_xlabel('Month')
ax1.set_ylabel('Temperature (°C)', color='tab:orange')
ax1.tick_params(axis='y', labelcolor='tab:orange')

# Create a secondary y-axis
ax2 = ax1.twinx()
ax2.plot(df['Month'], df['Humidity (%)'], marker='x', label='Humidity (%)', color='tab:blue')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Add title
plt.title('Dengue Outbreaks in Kolkata: Temperature vs Humidity')

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# Adjust layout
fig.tight_layout()

# Show plot
plt.show()
```
